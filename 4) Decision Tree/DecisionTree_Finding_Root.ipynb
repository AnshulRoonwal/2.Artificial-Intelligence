```python
#Convert the dataframe's 2nd column to hold correct values. This happens in place.
def convertDataType(df, columnIndex, toThisDataType, isString):
    for i in range(len(df)):
        if df[columnIndex][i] == '?' or df[columnIndex][i] == 'None': 
            if isString == True:
                df[columnIndex][i] = 'None'
            else:
                df[columnIndex][i] = 0.0
        else:
            df[columnIndex][i] = toThisDataType(df[columnIndex][i])
            
            
# convertDataType(results, 1, float, False)
# convertDataType(results, 13, int, True)
```
